```python
#cerner_2^5_2019
#Sample Pie chart using matplotlib for sample data
import matplotlib.pyplot as plt
colors = ['r','m','b','c']
portions = [2,7,5,11]
categories = ['A','B','C','D']

plt.pie(portions,
        labels=categories,
        colors=colors,
        startangle=30,
        shadow= True,
        explode=(0,0,0.08,0))

plt.title('Sample Pie Chart')
plt.show()
```
