```python
#cerner_2^5_2019
#Sample Stack Plots using Matplotlib for sample data
import matplotlib.pyplot as plt

days = [1,2,3,4,5]

A = [5,8,9,10,7]
B = [2,3,4,3,2]
C = [7,8,7,2,2]

plt.plot([],[],color='r', label='A', linewidth=5)
plt.plot([],[],color='b', label='B', linewidth=5)
plt.plot([],[],color='g', label='C', linewidth=5)


plt.stackplot(days, A,B,C, colors=['r','b','g'])

plt.xlabel('Days (x)')
plt.ylabel('Types (y)')
plt.title('Sample Stack Plot')
plt.legend()
plt.show()
```
