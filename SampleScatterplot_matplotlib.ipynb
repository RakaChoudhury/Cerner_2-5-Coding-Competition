```python
#cerner_2^5_2019
#Generate scatterplot using matplotlib for sample data
import matplotlib.pyplot as plt

x = [8,7,6,5,4,3,2,1,2]
y = [6,4,5,2,5,6,7,3,1]

plt.scatter(x,y, label='sample data', color='red', s=25, marker="o")

plt.xlabel('x')
plt.ylabel('y')
plt.title('Sample Scatterplot')
plt.legend()
plt.show()
```
